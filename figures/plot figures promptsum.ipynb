```python
# data to plot
n_groups = 3
means_pt = (0.32, 2.73, 5.67)
means_promptsum = (1.88, 13.42, 23.69)
means_gt = (11.8, 49.62, 70.31)

# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.2
opacity = 0.5

bs = plt.bar(index - 1 * bar_width, means_pt, bar_width,
alpha=opacity,
color='purple',
label='Prompt-tuning 100-shot')

dbs = plt.bar(index, means_promptsum, bar_width,
alpha=opacity,
color='royalblue',
label='PromptSum 100-shot')

gt = plt.bar(index + 1 * bar_width, means_gt, bar_width,
alpha=opacity,
color='limegreen',
label='Ground truth')

plt.xticks(index, ('1-grams', ' 2-grams', '3-grams'), fontsize=fontsize)
plt.ylim([0, 80])
plt.yticks(np.arange(0, 81, 20), fontsize=fontsize)
plt.title('Novel n-grams in summaries on CNN/DM (%)', fontsize=fontsize)

# plt.legend()
ax.legend(loc='upper left', prop={'size': 10})
plt.tight_layout()
fig.set_dpi(200)
plt.savefig("abstractiveness.png")
```
