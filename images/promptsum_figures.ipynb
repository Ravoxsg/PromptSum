```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
```

```python
fontsize = 8
```

```python
# data to plot
n_groups = 4

matplotlib.rc('font', size=8)
matplotlib.rc('axes', titlesize=8)

# cnndm
cnndm_finetune = (3.29, 16.25, 26.39, 33.48)
cnndm_soft = (1.11, 8.22, 15.57, 21.55)
cnndm_mix = (1.27, 10.42, 19.37, 26.29)
cnndm_target = (15.7, 54.51, 73.11, 81.92)

# xsum
xsum_finetune = (27.26, 73.51, 86.37, 90.28)
xsum_soft = (25.27, 74.03, 89.2, 94.2)
xsum_mix = (24.24, 72.55, 87.9, 93.09)
xsum_target = (35.52, 84.74, 95.92, 98.63)

# billsum
billsum_finetune = (13.61, 41.66, 57.72, 67.35)
billsum_soft = (6.05, 20.5, 31.81, 40.73)
billsum_mix = (7.88, 25.25, 37.6, 46.92)
billsum_target = (16.67, 48.16, 64.11, 73.11)

# samsum
samsum_finetune = (24.47, 64.39, 77.01, 84.02)
samsum_soft = (20.71, 60.67, 72.85, 81.13)
samsum_mix = (21.19, 62.83, 75.66, 84.18)
samsum_target = (34.12, 79.28, 90.51, 95.39)

# create plot
fig, axes = plt.subplots(2, 2, sharex=True)
((ax1, ax2), (ax3, ax4)) = axes
index = np.arange(n_groups)
bar_width = 0.15
opacity = 0.5

fig.suptitle('New n-grams (%)', fontsize=10)

# 1: CNNDM

ax1.set_title("CNN/DM", fontsize = fontsize)

ax1.bar(index, cnndm_finetune, bar_width,
alpha=opacity,
color='green',
label='Fine-tuning (PEGASUS-large)')

ax1.bar(index + bar_width, cnndm_soft, bar_width,
alpha=opacity,
color='blue',
label='Soft prompt-tuning (pre-trained)')

ax1.bar(index + 2 * bar_width, cnndm_mix, bar_width,
alpha=opacity,
color='purple',
label='PromptSum')

ax1.bar(index + 3 * bar_width, cnndm_target, bar_width,
alpha=opacity,
color='orange',
label='Ground-truth')

# 2: XSum

ax2.set_title("XSum", fontsize = fontsize)

ax2.bar(index, xsum_finetune, bar_width,
alpha=opacity,
color='green',
label='Fine-tuning (PEGASUS-large)')

ax2.bar(index + bar_width, xsum_soft, bar_width,
alpha=opacity,
color='blue',
label='Soft prompt-tuning (pre-trained)')

ax2.bar(index + 2 * bar_width, xsum_mix, bar_width,
alpha=opacity,
color='purple',
label='PromptSum')

ax2.bar(index + 3 * bar_width, xsum_target, bar_width,
alpha=opacity,
color='orange',
label='Ground-truth')

# 3: BillSum

ax3.set_title("BillSum", fontsize = fontsize)

ax3.bar(index, billsum_finetune, bar_width,
alpha=opacity,
color='green',
label='Fine-tuning (PEGASUS-large)')

ax3.bar(index + bar_width, billsum_soft, bar_width,
alpha=opacity,
color='blue',
label='Soft prompt-tuning (pre-trained)')

ax3.bar(index + 2 * bar_width, billsum_mix, bar_width,
alpha=opacity,
color='purple',
label='PromptSum')

ax3.bar(index + 3 * bar_width, billsum_target, bar_width,
alpha=opacity,
color='orange',
label='Ground-truth')

# 4: SAMSum

ax4.set_title("SAMSum", fontsize = fontsize)

ax4.bar(index, samsum_finetune, bar_width,
alpha=opacity,
color='green',
label='Fine-tuning (PEGASUS-large)')

ax4.bar(index + bar_width, samsum_soft, bar_width,
alpha=opacity,
color='blue',
label='Soft prompt-tuning (pre-trained)')

ax4.bar(index + 2 * bar_width, samsum_mix, bar_width,
alpha=opacity,
color='purple',
label='PromptSum')

ax4.bar(index + 3 * bar_width, samsum_target, bar_width,
alpha=opacity,
color='orange',
label='Ground-truth')

plt.setp(axes, 
    xticks = index + 1.5 * bar_width, 
    xticklabels = ['1-grams', ' 2-grams', '3-grams', '4-grams'],
    yticks = np.arange(0, 101, 20)
)

lines, labels = ax1.get_legend_handles_labels()
fig.legend(lines, labels, loc='lower center', prop={'size': 8}, bbox_to_anchor=(0.5,-0.1), ncol=2)

# plt.tight_layout()
fig.set_dpi(500)
plt.savefig("abstractiveness.png", bbox_inches='tight')
```
